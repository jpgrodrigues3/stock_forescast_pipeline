```python
# ── Divide fact_sales rows into all-year vs. seasonal ───────────────────────
from pyspark.sql import functions as F

# 1. Load fact_sales once
fact_sales = spark.table("Gold_Data.fact_sales.fact_sales")

# 2. Separate by the known seasonal key (60)
seasonal_fact_sales = fact_sales.filter(F.col("store_key") == 60)
all_year_fact_sales = fact_sales.filter(F.col("store_key") != 60)

# 3. Persist the two subsets as managed tables (Delta by default)
seasonal_fact_sales.write.mode("overwrite") \
    .saveAsTable("Machine_Learning.filtered_data.Seasonal_Fact_Sales")

all_year_fact_sales.write.mode("overwrite") \
    .saveAsTable("Machine_Learning.filtered_data.All_Year_Fact_Sales")

# # 4. Quick sanity check
# print(f"Seasonal rows  : {seasonal_fact_sales.count():,}")
# print(f"All-year rows  : {all_year_fact_sales.count():,}")
```
