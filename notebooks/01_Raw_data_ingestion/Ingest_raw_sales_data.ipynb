# Load sales Data

```python
from pyspark.sql.functions import col, expr
from pyspark.sql import SparkSession
from delta import *
```

```python
# Initialize SparkSession with Delta support
spark = SparkSession.builder \
    .appName("Raw City Data Ingestion") \
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension") \
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog") \
    .getOrCreate()
```

```python
sales_path = "Files/sales.csv" 
```

```python
# SPARK read parquet
df = spark.read.csv(sales_path, header = True)
print('Register the DataFrame as a SQL temporary view: sales')
df.createOrReplaceTempView('sales')
```

```python
# Display top 10 rows
print('Displaying top 10 rows: ')
df.show(10)
```

```python
# Grava como tabela registada
df.write.format("delta").mode("overwrite").saveAsTable("Bronze_data.sales.raw_sales_data")
```
