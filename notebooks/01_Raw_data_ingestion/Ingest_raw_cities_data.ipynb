# Load Cities Data

```python
from pyspark.sql.functions import col, expr
from pyspark.sql import SparkSession
from delta import *
```

```python
# Initialize SparkSession with Delta support
spark = SparkSession.builder \
    .appName("Raw City Data Ingestion") \
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension") \
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog") \
    .getOrCreate()
```

```python
cities_path = "Files/cities.csv" 
```

```python
# SPARK read parquet
df = spark.read.csv(cities_path, header = True)
print('Register the DataFrame as a SQL temporary view: Cities')
df.createOrReplaceTempView('cities')
```
