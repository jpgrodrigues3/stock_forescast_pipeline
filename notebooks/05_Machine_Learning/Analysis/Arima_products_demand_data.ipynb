````python
# ----------------------------------------------------------------------------------
# 0. Notebook config  – run once per session
# ----------------------------------------------------------------------------------
import mlflow
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

EXPERIMENT = "ARIMA_for_Products_w_demand_w_data"
FORECAST_HORIZON_WEEKS = 4          # how far ahead each model predicted
````

````python
# ----------------------------------------------------------------------------------
# 1. Load all runs + unpack metrics & product / store identifiers
# ----------------------------------------------------------------------------------
import mlflow
import pandas as pd
import numpy as np
import re

# Define your experiment name above, e.g.:
EXPERIMENT = "ARIMA_for_Products_w_demand_w_data"

# ── 1.1 Pull every run in the experiment ───────────────────────────────────────────
df = mlflow.search_runs(experiment_names=[EXPERIMENT])      # returns a Pandas DF

# ── 1.2 Convert every metric column to numeric ─────────────────────────────────────
metric_cols = [c for c in df.columns if c.startswith("metrics.")]
df[metric_cols] = df[metric_cols].apply(pd.to_numeric, errors="coerce")

# ── 1.3 Work out which column contains the run name ───────────────────────────────
run_name_col = next((c for c in [
                        "run_name",                 # used by AzureML SDK sometimes
                        "tags.mlflow.runName",      # MLflow default
                        "mlflow.runName",           # older templates
                        "tags.run_name"]           # custom pipelines
                     if c in df.columns), None)


# ── 1.4 Create product_id / store_id columns ──────────────────────────────────────
#  First try to grab them from tags; if missing, fall back to regex on the run name.
id_source = None
for p_tag, s_tag in [("tags.product_id", "tags.store_id"),
                     ("tags.product",    "tags.store"),
                     ("tags.Product",    "tags.Store")]:
    if p_tag in df.columns and s_tag in df.columns:
        df["product_id"] = pd.to_numeric(df[p_tag], errors="coerce").astype("Int64")
        df["store_id"]   = pd.to_numeric(df[s_tag], errors="coerce").astype("Int64")
        id_source = f"tags ({p_tag}, {s_tag})"
        break

if id_source is None:
    # No suitable tags found → parse IDs from the run name
    if run_name_col is None:
        raise ValueError("Cannot find run name column and no product/store tags exist; "
                         "please log tags or adjust the regex logic below.")
    pattern = re.compile(r"Product_(\d+)_Store_(\d+)", re.IGNORECASE)

    def extract_ids(name: str):
        m = pattern.search(str(name))
        return (int(m.group(1)), int(m.group(2))) if m else (pd.NA, pd.NA)

    df[["product_id", "store_id"]] = (
        df[run_name_col]
          .apply(lambda s: pd.Series(extract_ids(s), index=["product_id", "store_id"]))
          .astype("Int64")
    )
    id_source = f"parsed {run_name_col}"

print(f"✅ Identifier columns added using {id_source}.")
print(df[["run_id", run_name_col if run_name_col else '—', "product_id", "store_id"]].head())
print(f"{df['product_id'].nunique()} products × {df['store_id'].nunique()} stores represented "
      f"({len(df)} runs total).")
````

````python
# ----------------------------------------------------------------------------------
# 2. Aggregate performance by (product, store)
# ----------------------------------------------------------------------------------
agg = (df
       .groupby(["product_id","store_id"])
       .agg(aic_mean = ("metrics.aic",  "mean"),
            bic_mean = ("metrics.bic",  "mean"),
            mae_mean = ("metrics.mae",  "mean"),
            mape_mean= ("metrics.mape", "mean"),
            rmse_mean= ("metrics.rmse", "mean"),
            runs      = ("run_id",      "count"))
       .reset_index())

display(agg.head())

# Quick diagnostic plot – are some product-stores systematically worse?
plt.figure(figsize=(6,4))
plt.scatter(agg["mae_mean"], agg["mape_mean"], alpha=.4)
plt.xlabel("Average MAE"); plt.ylabel("Average MAPE")
plt.title("Forecast accuracy by product-store")
plt.show()


````

````python
# ----------------------------------------------------------------------------------
# 3. Pick the best run per (product, store) on a chosen metric
# ----------------------------------------------------------------------------------
SELECTION_METRIC = "metrics.aic"          # change to mae/mape/rmse if you prefer

# 3-1  Keep only runs that actually have the selection metric
df_valid = df[df[SELECTION_METRIC].notna()].copy()

if df_valid.empty:
    raise ValueError(f"No runs contain non-NA values for '{SELECTION_METRIC}'. "
                     "Check that the metric was logged.")

# 3-2  Sort so the best (lowest) metric is first, then drop duplicates
champions = (df_valid
             .sort_values(SELECTION_METRIC, ascending=True)
             .drop_duplicates(subset=["product_id", "store_id"], keep="first")
             .loc[:, ["product_id", "store_id", "run_id",
                      "metrics.aic", "metrics.bic" if "metrics.bic" in df.columns else None,
                      "metrics.mae", "metrics.mape", "metrics.rmse" if "metrics.rmse" in df.columns else None,
                      "artifact_uri"]]
             .reset_index(drop=True))

# 3-3  Remove any columns that were None placeholders
champions = champions.dropna(axis=1, how="all")

print(f"✅ Champion models selected: {len(champions)} "
      f"(one per product-store where '{SELECTION_METRIC}' was logged).")
champions.head()
````

````python
# ----------------------------------------------------------------------------------
# 2-B.  Average weekly sales per (product, store) 
# ----------------------------------------------------------------------------------
import pyspark.sql.functions as F

SALES_TABLE = "Machine_Learning.features.weekly_features"   # catalog name

sales_spark = (
    spark.read.table(SALES_TABLE)         
         .selectExpr(
             "cast(product_key as int)   as product_id",
             "cast(store_key   as int)   as store_id",
             "cast(weekly_quantities_sold as double) as quantity"
         )
)

avg_weekly_sales = (
    sales_spark.groupBy("product_id", "store_id")
               .agg(F.avg("quantity").alias("avg_weekly_sales"))
               .toPandas()
)

print(f"✅ avg_weekly_sales computed for {len(avg_weekly_sales)} product-store pairs.")
avg_weekly_sales.head()
````

````python
# ------------------------------------------------------------------
# 1. MAPE distribution across runs  (histogram with stats)
# ------------------------------------------------------------------
import matplotlib.pyplot as plt
import numpy as np

mape_vals = df["metrics.mape"].dropna()

fig, ax = plt.subplots(figsize=(7,4))
ax.hist(mape_vals, bins=30, edgecolor="black")
ax.set_title("MAPE distribution across runs", fontsize=14, pad=12)
ax.set_xlabel("MAPE (%)", fontsize=12)
ax.set_ylabel("Number of runs", fontsize=12)
ax.grid(axis="y", alpha=.3)

# ── overlay median & 90th percentile
median = np.median(mape_vals)
p90    = np.percentile(mape_vals, 90)
ax.axvline(median, color="red",  linestyle="--", label=f"Median = {median:.1f} %")
ax.axvline(p90,    color="green",linestyle="--", label=f"90th pct = {p90:.1f} %")
ax.legend()

plt.show()


# ------------------------------------------------------------------
# 2. MAE as % of weekly demand  (scatter + reference line)
# ------------------------------------------------------------------
# merge champion runs with average weekly sales
err = champions.merge(avg_weekly_sales, on=["product_id","store_id"])
err["mae_pct_of_demand"] = err["metrics.mae"] / err["avg_weekly_sales"] * 100

fig, ax = plt.subplots(figsize=(7,5))
ax.scatter(err["avg_weekly_sales"],
           err["mae_pct_of_demand"],
           alpha=.6, s=25)

ax.set_title("Forecast error vs. demand volume", fontsize=14, pad=12)
ax.set_xlabel("Average weekly demand (units)", fontsize=12)
ax.set_ylabel("MAE as % of demand", fontsize=12)
ax.grid(alpha=.3)

# reference bands
ax.axhline(20, color="orange", linestyle="--", linewidth=1,
           label="20 % error threshold")
ax.axhline(40, color="red",    linestyle="--", linewidth=1,
           label="40 % (poor)")

ax.legend(loc="upper right")
plt.show()
````

Primeiro gráfico:

**O que o grafico mostra**

1. **A maioria das barras encosta-se a esquerda, mas nao a zero**  
   – indica que muitas previsoes ate estao no intervalo correcto, mas ainda falham de forma apreciavel.

2. **A linha vermelha tracejada (mediana ≈ 94 %)**  
   – ordenamos todos os valores de MAPE; o do meio é 94 %.  
   → Metade das execucoes erra **ate** 94 % da procura real (ex.: prevê 19 quando o real e 10, ou 190 quando o real e 100).

3. **A linha verde tracejada (90.º percentil ≈ 270 %)**  
   – 90 % dos MAPEs ficam **abaixo** deste numero, logo so **1 em cada 10 execucoes** é pior do que 270 %.  
   → Nos piores casos o modelo chega a errar quase **3 ×** a procura real.

4. **Cauda longa a direita**  
   – alguns pares produto-loja vendem quase nada; qualquer erro faz disparar o MAPE e empurra a media para cima.

---

### Como se obtem os 94 % e os 270 %

```python
import numpy as np

p90    = np.percentile(mape_vals, 90)   # → 270 %
```

* **MAPE** e o erro medio absoluto em percentagem:  
   * Calculamos este valor para ~1680 execucoes → obtemos uma lista de percentagens.
   * **Mediana (94 %)**: valor central dessa lista.  
   * **90.º percentil (270 %)**: ponto abaixo do qual ficam 90 % dos valores.

````python
# 📦 Import necessary libraries
from pyspark.sql import functions as F
import matplotlib.pyplot as plt

# 🔗 Load the tables
fact_sales = spark.table("machine_learning.filtered_data.seasonal_fact_sales")
dim_stores = spark.table("Gold_data.dim_stores.dim_stores")

# 🧱 Join tables on store_key
joined_df = (
    fact_sales
    .join(dim_stores, on="store_key", how="left")
    .select(
        "store_key",
        "quantities_sold",
        "revenue",
        "price",
        "stock_quantity",
        "city_name",
        F.to_date(F.col("date_id").cast("string"), "yyyyMMdd").alias("date")
    )
)

# 📅 Extract week and year to group by week
joined_df = joined_df.withColumn("week_year", F.concat_ws("-", F.year("date"), F.weekofyear("date")))

# 📊 Aggregate sales per week
weekly_sales = (
    joined_df.groupBy("week_year")
    .agg(F.sum("quantities_sold").alias("total_sales"))
    .orderBy("week_year")
)

# 🐼 Convert to Pandas for plotting
pdf = weekly_sales.toPandas()

# 📈 Plot the weekly sales trend
plt.figure(figsize=(14, 6))
plt.plot(pdf["day_year"], pdf["total_sales"], marker="o")
plt.xticks(rotation=45, ha="right")
plt.xlabel("Day-Year")
plt.ylabel("Total Sales")
plt.title("Sales Trend for Seasonal Stores")
plt.grid(True)
plt.tight_layout()
plt.show()
````

````python
# 📅 Extract year and week number separately
weekly_sales_split = (
    joined_df.withColumn("year", F.year("date"))
             .withColumn("week", F.weekofyear("date"))
             .groupBy("year", "week")
             .agg(F.sum("quantities_sold").alias("total_sales"))
             .filter("year BETWEEN 2017 AND 2019")
             .orderBy("year", "week")
)

# 🐼 Convert to Pandas for plotting
pdf_split = weekly_sales_split.toPandas()

# 📈 Pivot to wide format: rows=week, columns=year, values=total_sales
pivot_df = pdf_split.pivot(index="week", columns="year", values="total_sales")

# 🖼️ Plot overlapping trends
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
for year in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[year], marker='o', label=str(year))

plt.xticks(range(1, 54), rotation=45)
plt.xlabel("Week Number")
plt.ylabel("Total Sales")
plt.title("Weekly Sales Trend for Seasonal Stores (2017–2019)")
plt.grid(True)
plt.legend(title="Year")
plt.tight_layout()
plt.show()
````

````python
# 📦 Import necessary libraries
from pyspark.sql import functions as F
import matplotlib.pyplot as plt
import pandas as pd
import datetime

# 🔗 Load the tables
fact_sales = spark.table("machine_learning.filtered_data.seasonal_fact_sales")
dim_stores = spark.table("Gold_data.dim_stores.dim_stores")

# 🧱 Join tables on store_key
joined_df = (
    fact_sales
    .join(dim_stores, on="store_key", how="left")
    .select(
        "store_key",
        "quantities_sold",
        "revenue",
        "price",
        "stock_quantity",
        "city_name",
        F.to_date(F.col("date_id").cast("string"), "yyyyMMdd").alias("date")
    )
)

# 📅 Extract year and month-day
joined_df = (
    joined_df
    .withColumn("year", F.year("date"))
    .withColumn("month_day", F.date_format("date", "MM-dd"))
)

# 📊 Aggregate daily sales by year and day (month-day)
daily_sales = (
    joined_df.groupBy("year", "month_day")
    .agg(F.sum("quantities_sold").alias("total_sales"))
    .filter("year BETWEEN 2017 AND 2019")
    .orderBy("month_day", "year")
)

# 🐼 Convert to Pandas
pdf = daily_sales.toPandas()

# 📐 Pivot to wide format
pivot_df = pdf.pivot(index="month_day", columns="year", values="total_sales")

# 🗓️ Get weekday info for each month_day using a fixed year (e.g. 2019)
# This works because day-of-week patterns repeat per date
pivot_df["weekday"] = [
    datetime.datetime.strptime(f"2019-{d}", "%Y-%m-%d").weekday()  # Monday=0, Sunday=6
    for d in pivot_df.index
]

# 🖼️ Plot with vertical weekend lines
plt.figure(figsize=(16, 6))
for year in [2017, 2018, 2019]:
    plt.plot(pivot_df.index, pivot_df[year], marker='o', label=str(year))

# Add vertical lines for weekends
for i, day in enumerate(pivot_df.index):
    if pivot_df.loc[day, "weekday"] in [5, 6]:  # Saturday or Sunday
        plt.axvline(x=day, color='yellow', linestyle='--', alpha=0.3)

plt.xticks(rotation=45, ha='right', fontsize=8)
plt.xlabel("Month-Day")
plt.ylabel("Total Sales")
plt.title("Daily Sales Trend with Weekends Highlighted (Seasonal Stores)")
plt.grid(True)
plt.legend(title="Year")
plt.tight_layout()
plt.show()

please upload to lakehouse
````
