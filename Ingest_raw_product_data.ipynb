# Load product Data

```python
from pyspark.sql.functions import col, expr
from pyspark.sql import SparkSession
from delta import *
```

```python
# Initialize SparkSession with Delta support
spark = SparkSession.builder \
    .appName("Raw City Data Ingestion") \
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension") \
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog") \
    .getOrCreate()
```

```python
product_path = "Files/product.csv" 
```

```python
# SPARK read 
df = spark.read.csv(product_path, header = True)
print('Register the DataFrame as a SQL temporary view: product')
df.createOrReplaceTempView('product')
```

```python
# Grava como tabela registada
df.write.format("delta").mode("overwrite").saveAsTable("Bronze_Data.product.raw_product_data")
```
